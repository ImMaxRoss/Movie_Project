```python
tn_budgets.info()
```

```python
tn_budgets
```

```python
tn_budgets['movie'].value_counts()
```

```python
full_movies_df = tn_budgets.copy()
```

```python
full_movies_df['worldwide_gross'] = full_movies_df['worldwide_gross'].str.replace('$','').str.replace(',','').astype(float)
full_movies_df['production_budget'] = full_movies_df['production_budget'].str.replace('$','').str.replace(',','').astype(float)
full_movies_df['domestic_gross'] = full_movies_df['domestic_gross'].str.replace('$','').str.replace(',','').astype(float)
```

```python
#Creating new DF and removing 0 from worldwide gross and production budget
full_movies_profit_df = full_movies_df.loc[full_movies_df['production_budget']!= 0].loc[full_movies_df['%_gross%']!=0]
```

```python

#creating profit column
full_movies_profit_df['profit'] = full_movies_profit_df['worldwide_gross'] - full_movies_profit_df['production_budget']
```

```python

#Creating profit margin column
full_movies_profit_df['profit_margin'] = (full_movies_profit_df['profit'] / full_movies_profit_df['worldwide_gross']) * 10
```
